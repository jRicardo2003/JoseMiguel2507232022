```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url ="https://raw.githubusercontent.com/jRicardo2003/JoseMiguel2507232022/refs/heads/main/data/clave_B_correlacion.csv"
df = pd.read_csv(url)
df.head()
```

Identificar las variables numéricas del dataset


```python
print("Columnas del dataset:")
print(df.columns)
#Para aplicar correlación es necesario identificar las variables numéricas presentes dentro del dataset, ya que este análisis únicamente puede realizarse entre valores cuantitativos.
```

```python
print("Tipos de datos:")
print(df.dtypes)
```

```python
variables_numericas = df.select_dtypes(include=['int64', 'float64'])

variables_numericas.head()
#Se seleccionaron únicamente las variables numéricas del dataset debido a que la correlación mide relaciones matemáticas entre variables cuantitativas.
#Estas variables serán utilizadas para calcular la matriz de correlación.
```

Revisar datos nulos, valores atípicos y consistencia general

```python
print("Valores nulos:")
print(df.isnull().sum())
```

```python
print("Duplicados:")
print(df.duplicated().sum())
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(data=variables_numericas)

plt.title("Detección de Valores Atípicos")

plt.xticks(rotation=45)

plt.show()
#Boxplot para detectar valores atípicos
```

Calcular la matriz de correlación

```python
correlation_matrix = variables_numericas.corr()

correlation_matrix
```

Crear un mapa de calor o gráfico de correlación

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Mapa de Calor de Correlación")

plt.show()
```

Identificar las correlaciones positivas más importantes

```python
#Durante el análisis se identificaron varias correlaciones positivas relevantes entre las variables del dataset.

#Las más importantes fueron:

# Frecuencia de compra y consumo total → 0.87
# Uso de la aplicación y consumo total → 0.52
# Frecuencia de compra y uso de la aplicación → 0.52
# Ticket promedio y consumo total → 0.51
# Ingresos mensuales y ticket promedio → 0.61
```

Identificar las correlaciones negativas más importantes

```python
# Correlaciones Negativas Más Importantes

#El análisis también permitió identificar relaciones negativas entre variables.

#Las correlaciones negativas más relevantes fueron:

#Satisfacción y reclamos → -0.77
# Satisfacción y tiempo de respuesta → -0.63
# Frecuencia de compra y tiempo de respuesta → -0.23
# Uso de la aplicación y tiempo de respuesta → -0.17

```

Explicar al menos 4 relaciones encontradas entre variables

```python
# Interpretación de Relaciones Encontradas

## Relación 1
#La frecuencia de compra y el consumo total presentan una correlación positiva fuerte de 0.87.
#Esto indica que los clientes que compran con mayor frecuencia tienden a generar un mayor consumo total para la empresa.



## Relación 2
#La satisfacción y los reclamos muestran una correlación negativa fuerte de -0.77.
#Esto significa que a medida que aumentan los reclamos, la satisfacción de los clientes disminuye considerablemente.



## Relación 3
#Los ingresos mensuales y el ticket promedio tienen una correlación positiva de 0.61.
#Esto sugiere que los clientes con mayores ingresos tienden a realizar compras de mayor valor económico.


## Relación 4
#El uso de la aplicación y el consumo total presentan una correlación positiva de 0.52.
# Esto indica que los clientes que utilizan más la aplicación generan mayores niveles de consumo dentro de la empresa.
```

Concluir cómo estas relaciones pueden ayudar a la empresa

```python
# Conclusiones

#El análisis de correlación permitió identificar relaciones importantes entre variables relacionadas con el comportamiento de los clientes.

#Los resultados muestran que variables como frecuencia de compra, uso de la aplicación y ticket promedio influyen directamente en el consumo total de los clientes.

#Asimismo, se identificó que factores como reclamos y tiempos de respuesta afectan negativamente la satisfacción del cliente.

#Esta información puede ayudar a la empresa a: Mejorar la atención al cliente, Optimizar tiempos de respuesta, Incentivar el uso de la aplicación, Implementar estrategias para aumentar la frecuencia de compra,Diseñar campañas enfocadas en clientes de mayor consumo.

#En conclusión, el análisis de correlación representa una herramienta útil para comprender el comportamiento de los clientes y apoyar la toma de decisiones estratégicas basadas en datos.
```
